# **3.Videogames released by platform on the basis of Release window**

```python
regions = ['NA', 'JP', 'EU', 'Other']
region_sales_sufix = '_Sales'    
df['Total_Sales'] = df['Global_Sales']
df['Total_Sales'] = df['Total_Sales'].fillna(0)
```

```python
tdf = df.copy()
tdf = df[df['Year_of_Release'].notna()]
tdf.sort_values('Year_of_Release',ascending=True)
```

```python
tdf.shape
```

```python
fig = px.histogram(tdf,animation_frame='Year_of_Release',x='Platform',y='Total_Sales',range_y=[0,600])
fig.update_xaxes(type='category')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

```python
top_tdf = tdf.groupby(['Platform','Year_of_Release']).agg({'Total_Sales':'count'}).reset_index()
top_tdf.columns = ['Platform','Year_of_Release','Count']
top_tdf = top_tdf[top_tdf['Count']>top_tdf['Count'].sum()* 0.01]
top_tdf['Year_of_Release'] = top_tdf['Year_of_Release'].astype(str)
```

```python
top_tdf.head()
```

```python
fig = px.bar(
    top_tdf,
    x='Platform',
    y='Count',
    color='Year_of_Release',
    barmode="group"
)
fig.update_layout(title="Total released video-games by best selling platforms on the basis of it's release window")
fig.update_xaxes(type='category')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

## **4. Sales Analysis**

```python
platform_tdf = tdf.groupby(['Platform', 'Year_of_Release']).agg({'Total_Sales': 'sum'}).reset_index()
platform_tdf = platform_tdf.sort_values('Year_of_Release', ascending=True)
platform_tdf.head()
```

```python
fig = px.bar(
    platform_tdf,
    x='Platform',
    y='Total_Sales',
    animation_frame='Year_of_Release',
    range_y=[0, 150],
)

fig.update_layout(title='Total Sales of Videogames on the basis of their release window')
fig.update_xaxes(type='category')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

```python
platform_top_tdf = platform_tdf[platform_tdf['Year_of_Release'].isin([2016, 2017, 2018, 2019])]
platform_top_tdf = platform_top_tdf[platform_top_tdf['Total_Sales'] > platform_top_tdf['Total_Sales'].sum() * 0.005]
platform_top_tdf['Year_of_Release'] = platform_top_tdf['Year_of_Release'].astype(str)

```

```python
fig = px.bar(
    platform_top_tdf,
    x='Platform',
    y='Total_Sales',
    color='Year_of_Release',
    barmode="group"
)
```

```python
fig.update_layout(title="Total sales by platforms (in Millions)")
fig.update_xaxes(type='category')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

```python
platform_sum_tdf = platform_tdf.groupby(['Platform']).agg({'Total_Sales': 'sum'}).reset_index()
platform_sum_tdf = platform_sum_tdf[platform_sum_tdf['Total_Sales'] > platform_sum_tdf['Total_Sales'].sum() * 0.03]
```

```python
fig = px.bar(platform_sum_tdf,x='Platform',y='Total_Sales')

fig.update_layout(title="Total sales of all time in the most important platforms (in Millions)")
fig.update_xaxes(type="category")
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

## **5.Distribution of sales**

```python
regions_agg = {}

for region in regions:
    regions_agg[region + region_sales_sufix] = 'sum'

regions_agg['Total_Sales'] = 'sum'
regions_agg
```

```python
geo_tdf = tdf.groupby(['Year_of_Release']).agg(regions_agg).reset_index()
geo_tdf = geo_tdf.sort_values('Year_of_Release', ascending=True)
geo_tdf.head(10)
```

```python
fig = go.Figure()

for region in regions:
    
    fig.add_trace(go.Scatter(
        x=geo_tdf['Year_of_Release'], 
        y=geo_tdf[region + region_sales_sufix], 
        mode='lines',
        name=region,
    ))
fig.update_layout(title="Total sales per year by region (in Millions)")
fig.update_xaxes(type='category')
fig.show()
```

```python
geo_year_df = tdf[["Year_of_Release",'NA_Sales','EU_Sales','JP_Sales','Other_Sales']]

geo_year_df[['NA_mean','EU_mean','JP_mean','Other_mean']] = geo_year_df.groupby('Year_of_Release')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].transform('sum')
geo_year_df = geo_year_df.drop(['NA_Sales','EU_Sales','JP_Sales','Other_Sales'], axis=1)
geo_year_df = geo_year_df.drop_duplicates()
geo_year_df = geo_year_df.sort_values("Year_of_Release")

temp_df1 = pd.DataFrame({'Place': ['NA_Sales']*geo_year_df.shape[0], 'Year':geo_year_df['Year_of_Release'], 'Sales': geo_year_df['NA_mean']})
temp_df2 = pd.DataFrame({'Place': ['EU_Sales']*geo_year_df.shape[0], 'Year': geo_year_df['Year_of_Release'], 'Sales': geo_year_df['EU_mean']})
temp_df3 = pd.DataFrame({'Place': ['JP_Sales']*geo_year_df.shape[0], 'Year': geo_year_df['Year_of_Release'], 'Sales': geo_year_df['JP_mean']})
temp_df4 = pd.DataFrame({'Place': ['Other_Sales']*geo_year_df.shape[0], 'Year': geo_year_df['Year_of_Release'], 'Sales': geo_year_df['Other_mean']})

final = pd.concat([temp_df1,temp_df2,temp_df3,temp_df4], axis=0)
final = final.sort_values("Year")

fig=px.bar(
    final,
    x='Place', 
    y="Sales", 
    animation_frame="Year",
    animation_group="Place", 
    color="Place", 
    hover_name="Place",
    range_y=[0, 200]
)
fig.update_layout(title="Year sales distribution by region",title_x=0.5)

fig.show()
```

## **6. Distribution of sales by genre**

```python
genre_tdf = tdf.groupby(['Genre']).agg(regions_agg)
genre_tdf = genre_tdf.sort_values('Total_Sales', ascending=False)
genre_tdf.head()
```

```python
genre_tdf.columns
```

```python
list(genre_tdf.index)
```

```python
# Reorder df to total genre scattewr plot
genre_total_tdf = genre_tdf.reset_index().sort_values('Total_Sales', ascending=False)
```

```python
fig = go.Figure()
    
fig.add_trace(go.Scatter(
    x=genre_total_tdf['Genre'], 
    y=genre_total_tdf['Total_Sales'], 
    mode='lines+markers',
))
fig.update_layout(title="Total sales by genre (Millions)")
fig.update_xaxes(type='category')
# fig.update_xaxes(categoryorder='total descending')
fig.show()
```

```python
genre_tops = list(genre_total_tdf.loc[genre_total_tdf['Total_Sales'] > genre_total_tdf['Total_Sales'].sum() * 0.03, 'Genre'])

genre_tops_df = tdf[tdf['Genre'].isin(genre_tops)]
fig  = go.Figure()
fig.add_trace(go.Pie(
    labels=genre_tops_df['Genre'], 
    values=genre_tops_df['Total_Sales'], 
    pull=[0, 0, 0.1, 0.05, 0, 0, 0.05, 0, 0.05],
))
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title="Percent of sales by Genre")
fig.show()
```
